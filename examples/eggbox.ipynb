```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

import nestle
```

```python
# Define the posterior density to be sampled:
tmax = 3.0 * np.pi
constant = np.log(1.0 / tmax**2)

def loglhood(t):
    return (2.0 + np.cos(t[0]/2.0)*np.cos(t[1]/2.0))**5.0

def prior(t):
    return t * 2.0 * tmax - tmax
```

```python
# plot the surface
t0, t1 = np.meshgrid(np.linspace(-tmax, tmax, 50),
                     np.linspace(-tmax, tmax, 50))
z = loglhood(np.array([t0, t1]))
plt.imshow(z, extent=(-tmax, tmax, -tmax, tmax))
```

```python
res = nestle.nest(loglhood, prior, 2, nobj=100, method='multi', verbose=True)
```
